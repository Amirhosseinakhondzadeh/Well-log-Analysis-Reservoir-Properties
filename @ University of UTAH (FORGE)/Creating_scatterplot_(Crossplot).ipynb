```python
import pandas as pd
import matplotlib.pyplot as plt
import lasio

las = lasio.read("/Users/amirhosseinakhondzadeh/Data/D5RL-00187_University of Utah_ME-ESW1_Run2_TCOM_Main.las")

df = las.df()
df.describe()
df.head(10)


""" Creating a Crossplot / Scatterplot """
""" Before we progress, we can set the default plot size for our scatterplots using plt.rcParams. """
plt.scatter(x='GR', y='NPHI', data=df)
plt.show()

""" We can see above that we now have a very simple and not very informative crossplot. Firstly, the values and
the way data is displayed is different to what we would expect. """
plt.rcParams['figure.figsize'] = (8, 8)

plt.scatter(x='GR', y='NPHI', data=df) # Set up the scatter plot
plt.xlim(0, 700) # Change the X and Y ranges
plt.ylim(-0.2, 1) # For the y axis, we need to flip by passing in the scale values in reverse order
plt.show()

""" Adding Labels to the Axes
The scatterplot above is not much use to anyone else, there are no labels or units on the axes. So we need to tell
the reader of the plot what is plotted against what.
We can add these in using plt.xlabel and plt.ylabel. """

plt.scatter(x='GR', y='NPHI', data=df) # Set up the scatter plot
plt.xlim(0, 700) # Change the X and Y ranges
plt.ylim(-0.2, 1) # For the y axis, we need to flip by passing in the scale values in reverse order
# Add in labels for the axes
plt.ylabel('Neutron Porosity (NPHI) - %', fontsize=14)
plt.xlabel('Gamma Ray (GR) - API', fontsize=14)
plt.show()


""" We can add a third variable onto our scatterplot through the use of colour. For this plot, we will add 
in the c argument and pass it the Gamma Ray (GR) column from the dataframe.
To control the range of colours shown we need to pass in values to vmin and vmax. In this example, 
we will set these to 0 and 100. """


# Set up the scatter plot
plt.scatter(x='GR', y='NPHI', data=df, c='GR', vmin=0, vmax=100)
# Change the X and Y ranges
plt.xlim(0, 700)
# For the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(-0.2, 1)

# Add in labels for the axes
plt.ylabel('Neutron Porosity (NPHI) - %', fontsize=14)
plt.xlabel('Gamma Ray (GR) - API', fontsize=14)
plt.show()


""" Changing Colormap and Adding Colorbar """
""" To understand what the colours on the plot mean, we can add a colorbar. There are a few ways to add colorbars to our plot. 
As we are just using plt.scatter which is a single figure, we can call upon plt.colorbar() and 
the also pass in the label we want to display alongside it.

To change the colour map we are using, we can set it to one of the ones at the webpage below using the cmap argument in
plt.scatter(). For this example, we will use rainbow. This will allow low Gamma Ray values to appear in purple/blue and
high values to appear in red. """

#Set the style sheet to bmh (gridline)
plt.style.use('bmh')

# Set up the scatter plot
plt.scatter(x='GR', y='NPHI', data=df, c='RHOZ', vmin=0, vmax=5, cmap='rainbow')

plt.xlim(0, 700) # Change the X and Y ranges
plt.ylim(-0.2, 1) # For the y axis, we need to flip by passing in the scale values in reverse order

# Add in labels for the axes
plt.ylabel('Neutron Porosity (NPHI) - %', fontsize=14)
plt.xlabel('Gamma Ray (GR) - API', fontsize=14)

# Make the colorbar show
plt.colorbar(label='Standard Resolution Formation Density {F11.4}')

plt.show()
```
