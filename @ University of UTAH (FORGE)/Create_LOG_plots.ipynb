## Logarithmic Data Visualization: Simple Log Plots

**Introduction:**

Logarithmic data visualization, known as "Simple Log Plots," is a powerful tool to visualize complex datasets that span a wide range of values. Unlike linear scales, logarithmic scales compress data exponentially, making it easier to identify patterns in datasets with extreme values.

**Key Points:**

Logarithmic plots use logarithmic scales for one or both axes, which reveal relationships and trends in data that might be obscured in linear plots.
These plots are useful in scientific research, finance, engineering, and other fields dealing with data covering multiple orders of magnitude.
Python's Matplotlib library provides an efficient way to create simple log plots using functions like plt.semilogx().
By incorporating logarithmic data visualization into data analysis workflows, we can make more informed decisions and gain deeper insights from diverse datasets.

**Working with Subplots in Matplotlib**

1) Next, we need to *create an axis on our figure object*. We can do this by **assigning ax1** is equal to plt.subplot2grid().

2) We first pass in the shape of the subplot grid, which for this example we will set 1 by 1. 
Next we specify the location of our subplot using index positions.

3) As we only have 1 subplot, we will set the location to zero zero. Rowspan and colspan are both set to 1, 
which means they will only be 1 column wide by 1 row high.

4) Next, we need to tell matplotlib what we want to plot. In this case we are going use ax1.plot, 
and pass in our Gamma Ray and Depth columns.

5) Using the ax notation, we can further customise the plot by setting the x and y limits, and displaying the grid.
fig = plt.subplots(figsize=(10,10))

```python
fig = plt.subplots(figsize=(10, 10))

# Set up the plot axis
ax1 = plt.subplot2grid((1, 1), (0, 0), rowspan=1, colspan=1)

ax1.plot("GR", "DEPTH", data=df, color="green")  # Call the data from the df dataframe
ax1.set_xlabel("Gamma")  # Assign a track title
ax1.set_xlim(0, 200)  # Change the limits for the curve being plotted
ax1.set_ylim(4700, 3500)  # Set the depth range
ax1.grid()  # Display the grid
plt.show()
```

**Adding the another Log**

To add a new track / subplot to our figure, we can repeat what we did above and add a new axis, ax2. 
We increment the second digit in the layout of plt.subplot2grid() so it is now 1 row, by 2 columns. 
For ax2, we need to place it in the second column and this is done by changing the location arguement from (0,0) to (0,1).

Also, as resistivity is normally scale logarithmically, we need to add in the line: ax2.semilogx().

```python
fig = plt.subplots(figsize=(10, 10))

# Set up the plot axes
ax1 = plt.subplot2grid((1, 3), (0, 0), rowspan=1, colspan=1)
ax2 = plt.subplot2grid((1, 3), (0, 1), rowspan=1, colspan=1)
ax3 = plt.subplot2grid((1, 3), (0, 2), rowspan=1, colspan=1)

ax1.plot("GR", "DEPTH", data=df, color="green")  # Call the data from the well dataframe
ax1.set_xlabel("Gamma")  # Assign a track title
ax1.set_xlim(0, 200)  # Change the limits for the curve being plotted
ax1.set_ylim(4700, 3500)  # Set the depth range
ax1.grid()  # Display the grid

ax2.plot("RHOZ", "DEPTH", data=df, color="red")
ax2.set_xlabel("Density")
ax2.set_xlim(1, 4)
ax2.semilogx()
ax2.set_ylim(4700, 3500)
ax2.grid()

ax3.plot("PEFZ", "DEPTH", data=df, color="blue")
ax3.set_xlabel("Formation Photoelectric Factor")
ax3.set_xlim(0, 8)
ax3.set_ylim(4700, 3500)
ax3.grid()

plt.show()
```

## Adding Fills to Log Plots

simple log plots

```python
fig = plt.subplots(figsize=(5, 8))

# Set up the plot axis
ax1 = plt.subplot2grid((1, 1), (0, 0), rowspan=1, colspan=1)

ax1.plot("GR", "DEPTH", data=df, color="black")  # Call the data from the df dataframe
ax1.set_xlabel("Gamma")  # Assign a track title
ax1.set_xlim(0, 200)  # Change the limits for the curve being plotted
ax1.set_ylim(4700, 4000)  # Set the depth range
ax1.grid()  # Display the grid
plt.show()
```

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

# Add fill_betweenx
plt.fill_betweenx(df['DEPTH'], 0, df['GR'], facecolor='green')

plt.xlim(0, 150)
plt.ylim(4400, 4000)
```

**fill_betweenx on Both Sides of Line**

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

#Add fill_betweenx to left
plt.fill_betweenx(df['DEPTH'], 0, df['GR'], facecolor='green')

#Add fill_betweenx to right
plt.fill_betweenx(df['DEPTH'], df['GR'], 150, facecolor='yellow')


plt.xlim(0, 150)
plt.ylim(4400, 4000)
```

**Fill To a Fixed Value**

*Defining a Cut-off*

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

# Using the where argument to fill to a fixed value
plt.fill_betweenx(df['DEPTH'], 80, df['GR'], where=df['GR']<=80, facecolor='yellow')
plt.fill_betweenx(df['DEPTH'], df['GR'], 80, where=df['GR']>=80, facecolor='green')


plt.xlim(0, 150)
plt.ylim(4400, 4000)
```

**Color Map Fills**

Variable Fill to the Left

```python
# Setup plot vlaues & color_index
left_value = 0
right_value = 150
span = abs(left_value-right_value)

cmap=plt.get_cmap('hot_r')

color_index = np.arange(left_value, right_value, span/100)
```

```python
color_index
```

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

plt.xlim(0, 150)
plt.ylim(4400, 4000)

for index in sorted(color_index):
    index_value = (index - left_value)/span
    color = cmap(index_value)
#     print(f'{index}\t{index_value}\t{color}')
    plt.fill_betweenx(df['DEPTH'], left_value, df['GR'], where=df['GR']>=index, color=color)
```

**Variable Fill to the Right**

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

plt.xlim(0, 150)
plt.ylim(4400, 4000)

for index in sorted(color_index):
    index_value = (index - left_value)/span
    color = cmap(index_value)
    
    plt.fill_betweenx(df['DEPTH'], right_value, df['GR'], where= df['GR']>=index, color = color)
```
