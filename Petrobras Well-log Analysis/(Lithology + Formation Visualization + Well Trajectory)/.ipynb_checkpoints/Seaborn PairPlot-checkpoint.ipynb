## Seaborn PairPlot

A Seaborn pairplot is a multivariate visualization that shows the pairwise relationships between all numeric variables in a dataset. It is a grid of plots, with each row and column representing a different variable. The diagonal plots show the distribution of each variable, while the off-diagonal plots show the relationship between two variables.

The Seaborn pairplot can be used to explore the relationships between variables, identify outliers, and visualize the distribution of data. It is a powerful tool for data exploration and can be used to gain insights into complex datasets.

In this part the **libraries** are imported.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Setting the stying of the Seaborn figure
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('/Users/amirhosseinakhondzadeh/CODE_WELLLOGS/Petrobras Well-log Analysis/Processed Data (out put of preprocessing == Input of ML)/df0_ML.csv')

# Remove high GR values to aid visualisation
df = df[df['GR']<= 200]
```

```python
df
```

<span style='color:gray'> <span style="font-size:20px;">**Creating First Pairplot of All Data**</span> </span>

```python
sns.pairplot(df)
```

Plotting Specific Columns

```python
cols_to_plot = ['NPHI', 'GR', 'PEFZ']
sns.pairplot(df[cols_to_plot])
plt.show()
```

Changing Diagonal to KDE

```python
sns.pairplot(df[cols_to_plot], diag_kind='kde')
plt.show()
```

<span style='color:gray'> <span style="font-size:20px;">**Adding a Regression Line**</span> </span>

```python
# Use plot_kws to change regression line colour
sns.pairplot(df[cols_to_plot], kind='reg', diag_kind='kde', plot_kws={'line_kws':{'color':'red'}})
plt.show()
```

<span style='color:gray'> <span style="font-size:20px;">**Splitting Data by Category**</span> </span>

```python
#sns.pairplot(df[cols_to_plot], hue='RHGX_HILT')
```

<span style='color:gray'> <span style="font-size:20px;">**Styling the Diagonal Plot**</span> </span>

```python
sns.pairplot(df[cols_to_plot], diag_kws={'color':'red'})
plt.show()
```

```python
sns.pairplot(df[cols_to_plot], diag_kws={'color':'red', 'bins':5})
plt.show()
```

<span style='color:gray'> <span style="font-size:20px;">**Styling the Points**</span> </span>

```python
sns.pairplot(df[cols_to_plot], diag_kws={'color':'red'}, plot_kws={'color':'green'})
plt.show()
```

```python
sns.pairplot(df[cols_to_plot], diag_kws={'color':'red'}, plot_kws={'color':'green', 's':5})
plt.show()
```

<span style='color:gray'> <span style="font-size:20px;">**Controlling Figure Size**</span> </span>

```python
sns.pairplot(df[cols_to_plot], height=2)
plt.show()
```
