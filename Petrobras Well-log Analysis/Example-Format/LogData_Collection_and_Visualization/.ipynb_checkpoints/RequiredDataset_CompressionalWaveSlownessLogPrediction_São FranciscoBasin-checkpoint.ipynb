```python
dlis0, *tail0 = dlis.load('/Users/amirhosseinakhondzadeh/Data/WorkPreparation/1-brsa-871-mg_8_tdd_cnl_ait.dlis')
#dlis0, *tail0 = dlis.load('/Users/amirhosseinakhondzadeh/Data/Forge 16A 78-32 TBEI image log with calibrated caliper arms_DLIS.dlis')
```

```python
dlis0.frames;
```

```python
for frame in dlis0.frames :
    
    #Search through the channels 
    for channel in frame.channels:
        if channel.name == frame.index:
            depth_units = channel.units
            
    print(f'Frame Name: \t\t {frame.name}')    
    print(f'Number of Channels: \t {len(frame.channels)}') 
    print(f'Depth Interval: \t {frame.index_min} - {frame.index_max} {depth_units}') 
    print(f'Depth Spacing: \t\t {frame.spacing} {depth_units}') 
    print(f'Channel Names: \t\t {str(frame.channels)}')                              # The individual logging curves 
    print('='*115)
```

<span style='color:deepskyblue'> <span style="font-size:19px;"> **STEP 2: create Pandas DataFrames containing all the required Well Log Data**

1. DEPTH .m = Measured Depth 

**Lithological Log**
1. GR .gAPI = Gamma-Ray

**Resistivity Log**
1. AT90 .ohm.m = Array Induction Two Foot Resistivity A90

**Porosity Logs**
1. RHOZ .g/cm3 = HRDD Standard Resolution Formation Density 
2. NPHI .m3/m3 = Thermal Neutron Porosity (Ratio Method)

**Sonic Log**
1. DTCO3    .us/ft   = Compressional Slowness from Measurement 3

```python
# Create variable "f" which stands for frame
f0 = dlis0.object('FRAME','60B') #RHGX_HILT, GR, NPHI
f1 = dlis0.object('FRAME','30B') #AT10, AT20, AT30, AT60, AT90
f2 = dlis0.object('FRAME','20B') #PEFZ
```

```python
# Create variables "c" which stands for channel
c0 = f0.curves()
c1 = f1.curves()
c2 = f2.curves()
```

```python
# Define the channels/logging curves of interest for each frame "f"
# Convert DEPTH values to metres, by multiplying by 0.00254. 
d0  = {'DEPTH' : c0['TDEP']*0.00254,
       'RHGX_HILT'  : c0['RHGX_HILT'],
       'GR'    : c0['GR'],
       'NPHI'  : c0['NPHI']}

d1  = {'DEPTH' : c1['TDEP']*0.00254,
       'AT10'  : c1['AT10'],
       'AT20'  : c1['AT20'],
       'AT30'  : c1['AT30'],
       'AT60'  : c1['AT60'],
       'AT90'  : c1['AT90']}

d2  = {'DEPTH' : c2['TDEP']*0.00254,
       'PEFZ'  : c2['PEFZ']}
```

```python
# Create dataframes containing the required well logging curves 
df0 = pd.DataFrame(d0)  # Depth spacing is 0.152344 m
df1 = pd.DataFrame(d1)  # Depth spacing is 0.076172 m
df2 = pd.DataFrame(d2)  # Depth spacing is 0.050781 m
```

```python
# Print the dataframes 
df0, df1, df2
```

<span style='color:green'> <span style="font-size:19px;"> **STEP 3: Data Pre-Processing**

Remove "-999.25" from each dataframe. It does not make any sense.

```python
df2.drop(df2.loc[df2['PEFZ']==-999.25].index, inplace=True)
```

```python
# Reset the Index of the dataframe 
df2 = df2.reset_index(drop=True)
```

Print the new dataframes 

```python
df0, df1, df2
```

Invert the order of values, in all the dataframes. By means "iloc indexer" we can select the columns in reverse order

```python
df0  = df0.iloc[::-1]
df1  = df1.iloc[::-1]
df2  = df2.iloc[::-1]
```

```python
# Reset the Index of each dataframe 
df0  = df0.reset_index(drop=True) 
df1  = df1.reset_index(drop=True)
df2  = df2.reset_index(drop=True)
```

```python
# Print the new dataframes 
df0, df1, df2
```

```python
df0.describe(), df1.describe(), df2.describe()
```

<span style='color:brown'> <span style="font-size:19px;"> **STEP 4: Create Well Log Plots**

```python
fig, axes = plt.subplots(figsize=(10, 10))
curve_names = ['GR(gAPI)', 'RHGX_HILT','NPHI', 'AT10(ohm.m)', 'AT20(ohm.m)', 'AT30(ohm.m)', 'PEFZ']

# genral title for all logs
#fig.suptitle("WELL 1-BRSA-871-MG", fontsize=20, fontweight='bold')
#fig.subplots_adjust(top=0.8, wspace=0.15)


#Set up the subplots grid
ax1 = plt.subplot2grid((1,4),(0,0),rowspan=1,colspan=1) 
ax2 = plt.subplot2grid((1,4),(0,1),rowspan=1,colspan=1)                                                               # HCAL log is the twin of GR log
ax3 = ax2.twiny()        
ax4 = plt.subplot2grid((1,4),(0,2),rowspan=1,colspan=1)         
ax5 = ax4.twiny()
ax6 = ax4.twiny()
ax7 = plt.subplot2grid((1,4),(0,3),rowspan=1,colspan=1)
###############
# ax1: First Track == GR

ax1.plot("GR", "DEPTH", data=df0, color="green", lw = 1)  # Call the data from the well dataframe
ax1.set_xlim(0, 200)  # Change the limits for the curve being plotted
ax1.spines['top'].set_edgecolor('green')

ax2.plot("RHGX_HILT", "DEPTH", data=df0, color="red", lw = 1)
ax2.set_xlim(1, 4)
# ax2.semilogx()
ax2.spines['top'].set_edgecolor('red')
ax2.spines['top'].set_position(('outward',0))

ax3.plot("NPHI", "DEPTH", data=df0, color="blue", lw = 1)
ax3.set_xlim(-0.1, 0.2)
ax3.spines['top'].set_edgecolor('blue')
ax3.spines['top'].set_position(('outward',50))

# Fifth Track: AT10
ax4.plot("AT10","DEPTH", data= df1, color="lime", lw = 1)
ax4.set_xlim(0.7,5000)   
ax4.semilogx()   
ax4.spines['top'].set_edgecolor('lime')
ax4.spines['top'].set_position(('outward',0))

# Fifth Track: AT20
ax5.plot("AT20","DEPTH", data= df1, color="yellow", lw = 1)
ax5.set_xlim(0.7,5000)                                                   # [0.2-2000] is the typical rang
ax5.semilogx()
ax5.spines['top'].set_edgecolor('yellow')
ax5.spines['top'].set_position(('outward',50))

# Sixth Track: AT30
ax6.plot("AT30","DEPTH", data= df1, color="brown", lw = 1)
ax6.set_xlim(0.7,5000)
ax6.semilogx()
ax6.spines['top'].set_edgecolor('brown')
ax6.spines['top'].set_position(('outward',100))

# Seventh Track: PEFZ
ax7.plot("PEFZ","DEPTH", data= df2, color="black", lw = 1)
ax7.set_xlim(0,7.5)
ax7.spines['top'].set_edgecolor('black')



# Set up the common elements between the subplots
for i, ax in enumerate(fig.axes):
    ax.set_ylim(3450, 3200)  # Set the depth range

    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.set_xlabel(curve_names[i])
    ax.grid()
    if i == 4:
        ax.spines["top"].set_position(("axes", 1.08))
    else:
        ax.grid()

# Hide tick labels on the y-axis
for ax in [ax2, ax3, ax4, ax5, ax6, ax7]:
    plt.setp(ax.get_yticklabels(), visible=False)

plt.show()
```

<span style='color:brown'> <span style="font-size:19px;"> **STEP 5: create an unique DataFrame containing all the Well Log Data required for the ML prediction**

```python
# Perform "re-sampling"
# We decide that the "depth spacing", for each dataframe, should be 0.1524m 
df1 = df1[df1.index % 2 == 0].reset_index(drop=True) # We keep only the even rows + reset the Index
df2 = df2[df2.index % 3 == 0].reset_index(drop=True) # We keep only the rows whose Index is divisible by 3 + reset the Index 
```

```python
# Print the new dataframes 
df0, df1, df2
```

Concatenate, horizontally, the new dataframes 

```python
df0_ML = pd.concat([df0['DEPTH'],df0['GR'],df0['RHGX_HILT'],df0['NPHI'],df1['AT10'],df1['AT20'],df1['AT30'],df1['AT60'],df1['AT90'],df2['PEFZ']], axis = 1, ignore_index = False)
df0_ML
```

We drop the NaN values. The new depth interval is [3241.243164-3415.741211] (m)

```python
df0_ML = df0_ML.dropna()
df0_ML
```

Convert the DataFrame into a CSV file (CSV file is required for ML Algorithms implementation)

```python
df0_ML.to_csv('df0_ML.csv', index=None)  # df0_ML.csv is the name of the csv file
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,20))
fig.suptitle("WELL 1-BRSA-948-MG", fontsize=42, fontweight='bold')
fig.subplots_adjust(top=0.8, wspace=0.15)

track_names = ['GR(gAPI)','NPHI(m3/m3)','DTCO(us/ft)']


#Set up the subplots grid
ax1 = plt.subplot2grid((1,3),(0,0),rowspan=1,colspan=1)                                                               # HCAL log is the twin of GR log

ax2 = plt.subplot2grid((1,3),(0,1),rowspan=1,colspan=1)         

ax3 = plt.subplot2grid((1,3),(0,2),rowspan=1,colspan=1)

###############################################################################################################################################################################

# First Track: GR
ax1.plot("GR","DEPTH", data = df3, color = "lime", lw = 1)            
ax1.set_xlim(0,50)                                                      
ax1.spines['top'].set_position(('outward',0)) 
ax1.set_xlabel('GR(gAPI)',color='lime')  
ax1.set_ylabel('Measured Depth(m)',color='black') 
ax1.tick_params(axis='x', colors='lime')
ax1.spines["top"].set_edgecolor("lime")

# Third Track: NPHI
ax2.plot("NPHI","DEPTH", data= df3, color="blue", lw = 1)
ax2.set_xlim(0.45,-0.00015)
ax2.spines['top'].set_position(('outward',0))
ax2.set_xlabel('NPHI(m3/m3)', color='blue') 
ax2.tick_params(axis='x', colors='blue')
ax2.spines["top"].set_edgecolor("blue")

# Fourth Track: DTCO
ax3.plot("DTCO","DEPTH", data= df3, color="blue", lw = 1)
ax3.set_xlim(60,50)                                                      
ax3.spines['top'].set_position(('outward',0))
ax3.set_xlabel('DTCO(us/ft)', color='blue') 
ax3.tick_params(axis='x', colors='blue')
ax3.spines["top"].set_edgecolor("blue")

###############################################################################################################################################################################

for i, ax in enumerate(fig.axes):
    
    ax.set_ylim(3605,3405.8)                            # Set the depth range
    #ax.set_ylim(,)                                   # Depth interval for the prediction
    
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.set_xlabel(track_names[i])
    
    if i == 3:                                            
        ax.spines["top"].set_position(("axes", 1.08))
    else:
        ax.grid(color='black', lw=0.7)                # To display the grid 
 

# Add a black border to each subplot 
for ax in [ax1,ax2,ax3]:
    ax.spines['bottom'].set_color('black')
    ax.spines['left'].set_color('black')
    ax.spines['right'].set_color('black') 


# Hide tick labels on the y-axis
for ax in [ax2,ax3]:
    plt.setp(ax.get_yticklabels(), visible=False)
    
# Reduce the space between each subplot
fig.subplots_adjust(wspace = 0.2)
```
