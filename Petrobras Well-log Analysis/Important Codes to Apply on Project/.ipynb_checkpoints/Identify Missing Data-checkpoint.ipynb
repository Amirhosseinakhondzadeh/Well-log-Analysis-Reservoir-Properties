# ==============================================================

count the number of missing (null) values in each column of a DataFrame called df. It calculates and returns a Series object that shows the count of missing values for each column in the DataFrame.

```python
df.isnull().sum()
```

filter rows in a DataFrame (df) where at least one value in any column of that row is missing (null or NaN).

```python
df[df.isnull().any(axis=1)]
```

Drop Missing Data

```python
# Drop rows containing NaN
df.dropna()
```

```python
#Drop columns containing NaN
df.dropna(axis=1)
```

Fill the NaNs

```python
# Fill with a fixed value
df.fillna(1000)
```

"This code is used to fill in missing values in the DataFrame 'df' by propagating the previous non-null value forward. In other words, it replaces each missing value with the most recent non-null value that occurred before it in the same column."

```python
# Fill Forward
df.fillna(method='ffill')
```

"This code is used to fill in missing values in the DataFrame 'df' by propagating the next non-null value backward. In other words, it replaces each missing value with the first non-null value that occurs after it in the same column."

```python
# Fill Backward
df.fillna(method='bfill')
```

"This code is used to estimate and fill in missing values in the DataFrame 'df' by using linear interpolation. Linear interpolation calculates values that fit within the missing gaps based on the values of adjacent data points, creating a smooth linear progression between known data points."

```python
# Interpolate values
df.interpolate(method='linear')
```

"This code is used to replace missing values in the DataFrame 'df' with the mean (average) value of each column. It calculates the mean separately for each column and uses those calculated means to fill in the missing values in the corresponding columns."

```python
#Fill with mean
df.fillna(value=df.mean())
```
